```python
#Removing % from in_rate column
chargedoff_loans["int_rate"] = chargedoff_loans['int_rate'].apply(lambda x: float(x.split("%")[0]))
```

```python
chargedoff_loans.int_rate.value_counts()
```

```python
# Employment Lengh - Converting column from String type to Number so that this can be used for the analysis

chargedoff_loans['emp_length'] = chargedoff_loans['emp_length'].astype("str")
chargedoff_loans['emp_length'] = chargedoff_loans['emp_length'].apply(lambda x: x.replace("years",""))
chargedoff_loans['emp_length'] = chargedoff_loans['emp_length'].apply(lambda x: x.replace("year",""))
chargedoff_loans['emp_length'] = chargedoff_loans['emp_length'].apply(lambda x: x.replace("10+","11"))
chargedoff_loans['emp_length'] = chargedoff_loans['emp_length'].apply(lambda x: x.replace("< 1","0"))
```

```python
chargedoff_loans.emp_length.value_counts()
```

```python
#removing missing values from 'emp_length' column
chargedoff_loans = chargedoff_loans[~(chargedoff_loans['emp_length']=="nan")]
```

```python
chargedoff_loans.emp_length.value_counts()
```

```python
#Converting emp_length to integer
chargedoff_loans['emp_length']=chargedoff_loans['emp_length'].astype('int')
```

## Converting Continuous variable to Categorical

```python
# Column = annual_inc
# Categorizing Annual Income in Three Categories as 
#    1. LOW
#    2. MEDIUM
#    3. HIGH 
    
chargedoff_loans["annual_inc"]
```

```python
chargedoff_loans["annual_inc"].describe()
```

```python
#Low-income: 0-50000
#Mid-income: 50001-100000
#High-income: 100001>
def annual_inc_conversion(x):
  if x<= 50000:
    return "Low Income"
  elif x> 50000 and x <= 100000:
    return "Medium Income"
  else:
    return "High Income"
```

```python
chargedoff_loans["annual_inc"] = chargedoff_loans["annual_inc"].apply(lambda x:annual_inc_conversion(x))
```

```python
chargedoff_loans.annual_inc.value_counts()
```

```python
chargedoff_loans['annual_inc'] = chargedoff_loans['annual_inc'].astype("str")
```

```python
chargedoff_loans['annual_inc']
```

```python
#low-intersrt: 7.5
#mid-interst: 7.6-13
#high-insterst: 13+
def interst_rate_conversion(x):
  if x<= 7.5:
    return "Low Interst Rate"
  elif x>7.5  and x <= 13.0:
    return "Medium Interst Rate"
  else:
    return "High Interst Rate"
```

```python
chargedoff_loans["int_rate"]= chargedoff_loans["int_rate"].apply(lambda x:interst_rate_conversion(x))
```

## Univariate Analysis

```python
#Function to count values for a given column

def cat_analysis(x,):
  print(chargedoff_loans[str(x)].value_counts())
```

```python
# Function to plot graph using a given column

def univarite_categorical_analysis(column):
  sns.countplot(x=chargedoff_loans[str(column)])
```

```python
#Interest Rate
cat_analysis('int_rate')
```

Recommendation: Loan with High Interest Rate are Likely to Default

```python
univarite_categorical_analysis("int_rate")
```

```python
#Purpose
cat_analysis('purpose')
```

Recommendation: If Purpose of the loan is Debt COnsolidation, Credit Card or not specified the purpose (Other) then its likely to default

```python
plt.figure(figsize=(23,10))
univarite_categorical_analysis("purpose")
plt.show()
```

## Bi-Variate Analysis

```python
def cat_data_visualization(x, y):
  sns.countplot(data=chargedoff_loans, x=str(x), hue=str(y))
```

```python
chargedoff_loans.groupby(['int_rate', 'term']).size().reset_index(name='Count').sort_values('Count',ascending=False)
```

Recommendation: Though Loans with High Interest & High Tenure need likely to Default; however loan with 36 months of tenure have observed Defaulted. So company should offer longer term for the repayment.

```python
cat_data_visualization("int_rate", "term")
```

```python
chargedoff_loans.groupby(['annual_inc', 'int_rate']).size().reset_index(name='Count').sort_values('Count',ascending=False)
```

Recommendation: There are two categories which has high risk of Default
                1. Application with Medium Income and ready to take loan at High Interest Rate
                2. Application with Low Income and ready to take loan at High Interest Rate

```python
cat_data_visualization("annual_inc", "int_rate")
```

```python
chargedoff_loans.groupby(['home_ownership', 'int_rate']).size().reset_index(name='Count').sort_values('Count',ascending=False)
```
